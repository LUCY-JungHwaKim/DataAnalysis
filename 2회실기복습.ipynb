```python
x_train = pd.read_csv('./yemoonsaBigdata/datasets/Part3/204_x_train.csv')
x_test = pd.read_csv('./yemoonsaBigdata/datasets/Part3/204_x_test.csv')
y_train = pd.read_csv('./yemoonsaBigdata/datasets/Part3/204_y_train.csv')
```

```python
x_train.head()
```

```python
x_train = x_train.iloc[:, 1:]
x_test = x_test.iloc[:, 1:]
```

```python
y_train = y_train.iloc[:, 1:]
```

```python
x_train.info()
```

```python
X_data = pd.concat([x_train, x_test])
```

```python
from sklearn.preprocessing import LabelEncoder

lhe = LabelEncoder()

COL_CAT = ['Warehouse_block', 'Mode_of_Shipment', 'Product_importance', 'Gender']

for _col in COL_CAT:
  lhe.fit(X_data[_col])
  X_data[_col] = lhe.transform(X_data[_col])
```

```python
X_trains = X_data.iloc[:len(x_train)].reset_index(drop=True)
X_tests = X_data.iloc[len(x_train):].reset_index(drop=True)
X_trains
```

```python
from sklearn.preprocessing import MinMaxScaler

scl = MinMaxScaler()

scl.fit(X_trains)
X_trains_scl = scl.transform(X_trains)
X_tests_scl = scl.transform(X_tests)
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(X_trains_scl, y_train, test_size=0.3)
X_tr.shape
```

```python
X_val.shape, y_tr.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
y_tr_fin = y_tr['Reached.on.Time_Y.N']
```

```python
model_rf = RandomForestClassifier(n_estimators=10)
model_rf.fit(X_tr, y_tr_fin)
```

```python
y_pred = model_rf.predict_proba(X_val)[:, 1]

from sklearn.metrics import roc_auc_score

scr = roc_auc_score(y_val, y_pred)
scr
```

```python
model_xgb = XGBClassifier()
model_xgb.fit(X_tr, y_tr_fin)
y_pred_xgb = model_xgb.predict_proba(X_val)[:, 1]
scr_xgb  = roc_auc_score(y_val, y_pred_xgb)
scr_xgb
```
