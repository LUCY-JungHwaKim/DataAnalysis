## 제 2유형

```python
X_train = pd.read_csv('./yemoonsaBigdata/datasets/Part3/304_x_train.csv', index_col=0)
X_test = pd.read_csv('./yemoonsaBigdata/datasets/Part3/304_x_test.csv', index_col=0)
y_train = pd.read_csv('./yemoonsaBigdata/datasets/Part3/304_y_train.csv', index_col=0)
```

```python
X_train.info()
```

```python
X = pd.concat([X_train, X_test])
X.info()
```

```python
X.describe()
```

```python
## 독립형 변수 중 범주형 변수 인코딩하기
## 동일한 기준을 적용하기 위해 한번에 레이블링 인코딩 수행
from sklearn.preprocessing import LabelEncoder

lhe = LabelEncoder()

COL_CAT = ['Employment Type', 'GraduateOrNot', 'FrequentFlyer', 'EverTravelledAbroad' ]

for col in COL_CAT:
  lhe.fit(X[col])
  X[col] = lhe.transform(X[col])
```

```python
X_trains = X[:1490]
X_tests = X[1490:]
```

```python
X_trains.shape
```

```python
X_trains
```

```python
X_train.tail()
```

```python
y_train.tail()
```

```python
## scaler 진행 후 데이터셋 분리
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_trains)
X_tr_scl = scaler.transform(X_trains)
X_test_scl = scaler.transform(X_tests)
```

```python
## 데이터셋 분리
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(X_tr_scl, y_train, test_size=0.3)
```

```python
X_tr.shape, X_val.shape
```

```python
X_tr_scl
```

```python
y_tr_fin = y_tr['TravelInsurance']
y_val_fin = y_val['TravelInsurance']
```

```python
### 모델 학습
## 보험 가입 여부 예측 --> 분류 모델 사용
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import roc_auc_score, classification_report
## score 넣어도됨
def get_scores(model, val_x, val_y):
  #print("hello")
  X_val_pred = model.predict_proba(val_x)[:,1]
  result = roc_auc_score(val_y, X_val_pred)
  #print(result)
  return result
```

```python
def model_trains(train_x, train_y, val_x, val_y):
  ## 랜덤포레스트 모델
  est_lst = [10, 20, 30, 40, 50]
  for i in range(len(est_lst)):
    model_rf = RandomForestClassifier(n_estimators=est_lst[i]).fit(train_x, train_y)
    print("model_rf + est" , (str(est_lst[i])))
    print(get_scores(model_rf, val_x, val_y))
  ## 의사결정트리 모델
  depth_lst = [2,3,4,5,6,7,8]
  for i in range(len(depth_lst)):
    model_dt = DecisionTreeClassifier(max_depth=depth_lst[i]).fit(train_x, train_y)
    print("model_df + depth" , (str(depth_lst[i])))
    print(get_scores(model_dt, val_x, val_y))

  
  ## XGB 
  model_xgb = XGBClassifier().fit(train_x,train_y)
  print('model_xgb', get_scores(model_xgb, val_x, val_y))  

  ## LogisticRegression
  model_lr = LogisticRegression().fit(train_x, train_y)
  print('model_lr', get_scores(model_lr, val_x, val_y))  



```

```python
model_trains(X_tr, y_tr_fin, X_val, y_val_fin)
```

```python
best_model = XGBClassifier().fit(X_tr, y_tr_fin)
print(get_scores(best_model, X_val, y_val_fin))
y_val_label = best_model.predict(X_val)
print(classification_report(y_val_label, y_val_fin))
```

```python
## 테스트 실행
y_test_pr = best_model.predict_proba(X_test_scl)[:, 1]
y_test_pr
```

```python
result = pd.DataFrame({'pred':y_test_pr})
result.head()
# result.to_csv()
```
