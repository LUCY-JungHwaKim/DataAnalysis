```python
#문제 5. sepal_length 변수가 이상치를 가지는 데이터 행수 계산
# 이상치 범위는 (평균 + 2*표준편차) 이상이거나 (평균 - 2*표준편차) 이하
```

```python
iris.info()
```

```python
avg_sl = iris.sepal_length.mean()
std_sl = iris.sepal_length.std()
```

```python
iqr1 = avg_sl + (2*std_sl)
iqr2 = avg_sl - (2*std_sl)
```

```python
len(iris[(iris.sepal_length >= iqr1) | (iris.sepal_length <= iqr2)])
```
