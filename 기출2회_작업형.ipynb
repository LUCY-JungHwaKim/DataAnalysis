```python
data_thir = pd.read_csv('./yemoonsaBigdata/datasets/Part3/203_housing.csv')
```

```python
data_thir.info()
```

```python
## 그냥 확인용
data_thir.population.describe()
```

```python
## 상위10% = quantile(0.9)
```

```python
data_thir.population.quantile(0.75) ## 상위 25%라는 뜻
```

```python
std_pop = data_thir['population'].std()
print(std_pop)
```

```python
std_avg = data_thir['population'].mean()
```

```python
iqr1 = std_avg - (1.5 * (std_pop))
iqr2 = std_avg + (1.5 * (std_pop))
## 이상값은 평균에서 1.5*표준편차를 초과하거나 미만인 값의 범위로 정함
```

```python
result = data_thir[(data_thir['population'] > iqr2) | (data_thir['population'] < iqr1)].population.sum()
print(result)
```
