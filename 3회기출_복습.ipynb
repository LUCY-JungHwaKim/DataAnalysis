```python
import pandas as pd
X_tr= pd.read_csv('./yemoonsaBigdata/datasets/Part3/304_x_train.csv', index_col=0)
X_ts= pd.read_csv('./yemoonsaBigdata/datasets/Part3/304_x_test.csv', index_col=0)
y_tr= pd.read_csv('./yemoonsaBigdata/datasets/Part3/304_y_train.csv', index_col=0)
```

```python
y_tr.info()
```

```python
X = pd.concat([X_tr, X_ts])
```

```python
## 레이블링 인코딩
## X 전체 합쳐서 레이블링 인코딩하기
from sklearn.preprocessing import LabelEncoder

lhe = LabelEncoder()

COL_CAT = ['Employment Type', 'GraduateOrNot', 'FrequentFlyer', 'EverTravelledAbroad']

for _col in COL_CAT:
  lhe.fit(X[_col])
  X[_col]= lhe.transform(X[_col])
```

```python
X_train = X.iloc[:len(X_tr)]
X_test = X.iloc[len(X_tr):]
X_train.info(), X_test.info()
```

```python
## 스켈링 진행
## 스켈링 할때는 train으로 학습하고 train,test 둘다 transform 하기
from sklearn.preprocessing import MinMaxScaler

scl = MinMaxScaler()

scl.fit(X_train)
X_train_scl = scl.transform(X_train)
X_test_scl = scl.transform(X_test)
```

```python
X_train_scl
```

```python
len(y_tr)
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(X_train_scl, y_tr, test_size=0.3)

X_val.shape
```

```python
y_tr
```

```python
y_tr_fin = y_tr['TravelInsurance']
y_val_fin = y_val['TravelInsurance']
y_tr_fin ## 요렇게 정해주고 하는게 나을듯
```

```python
## 학습 실행
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
model_rf = RandomForestClassifier(n_estimators = 10)

model_rf.fit(X_tr, y_tr_fin)

y_pred = model_rf.predict_proba(X_val)[:, 1]

score_rf = roc_auc_score(y_val, y_pred) ## 순서 기억하기 val, pred
score_rf
```

```python
pred = model_rf.predict_proba(X_test_scl)[:,1]
pred
```

```python
result = pd.DataFrame({'y_pred':pred})
result
```

```python
result.to_csv('./result.csv', index=True)
```
